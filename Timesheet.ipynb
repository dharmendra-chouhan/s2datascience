**Clockify and Bugzilla Merging Data**

```
#@title  {display-mode: "form"}
scriteria = bugs20200608['Bug ID'].values.tolist()  
df=pd.DataFrame(clockify20200608)
clockifybugs = df[0:0]
clockify20200608['Bug ID'] =0
for x in scriteria:
  df=pd.DataFrame(clockify20200608[clockify20200608['Description'].str.upper().str.contains(str(x))])
  df['Bug ID'] =x
  clockifybugs=clockifybugs.append(df)
clockifybugs = pd.merge(clockifybugs, bugs20200608, on='Bug ID')
clockifybugsmer=clockifybugs[['Bug ID','Product', 'User','Assignee','Description','Summary','Orig. Est.','Duration (decimal)','Status','Resolution','Actual Hours','Start Date_x','End Date']] 
clockifybugsmer
```

Clockify and Bugzilla Bug,User,start and end date wise Duration sum

```
import numpy as np
grouped=clockifybugsmer.groupby(['User','Bug ID','Start Date_x','End Date'])
print(grouped['Duration (decimal)'].agg([np.sum]))
```
